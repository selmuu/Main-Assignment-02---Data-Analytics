### 1.2

```python
matrix = np.random.random(size=(10, 10))
print(matrix.shape)
print(matrix.dtype)
```

### 1.3

```python
print(np.mean(matrix))
print(matrix.sum())
print(matrix.min())
print(matrix.max())
```

### 1.4

```python
print(np.mean(matrix[0]))
print(matrix[0].sum())
print(matrix[0].min())
print(matrix[0].max())
```

### 1.5

```python
rolls = np.random.randint(1,7, size=100)
counts = np.bincount(rolls)

print("The number of times each number rolled on a 100 dice rolls")
for num in range(1, 7):
    print(f"{num}: {counts[num]}")
```
